## 결측치

```python
import pandas as pd
```

```python
age = pd.Series([ None, 22, 24, 26, 28, 30, 32, 34, 36, 38 ])
score = pd.Series([2.4, 2.6, 3.0, 3.4, None, 4.1, 4.3 ,4.5, 3.9, 4.2])
salary = pd.Series([2400, 2600, 2700, 2800, 2900, 3000, 3200, None, 3800, 4000])
stu_class = pd.Categorical([1,2,3,4,1,1,2,2,3,4])
gender = pd.Categorical(['F','M','F','M','F','F','M','M','F','M'])

df = pd.DataFrame({'age' : age,
                   'score' : score,
                   'salary' : salary,
                   'class' : stu_class,
                   'gender' : gender})


df
```

```python
df = pd.DataFrame({'age' : age,
                   'score' : score,
                   'salary' : salary,
                   'class' : stu_class,
                   'gender' : gender})

df

# df 복제 

df_new = df.copy()
df_new
```

### * 결측치 확인

```python
pd.isna(df_new)
# 결측치만 true 로 나오고 입력된 값들은 false 로 나옴
```

```python
# 열별로 결측치 갯수 확인 
df_new.isnull().sum() 
```

### * 결측치 제거

```python
#결측치를 가지고 있는 행들을 삭제

df_new = df_new.dropna()

# 결측치를 가지고 있는 행 중 원하는 열만 삭제

# df_new = df_new['age'].dropna()
 
df_new
```

```python
#결측치 해당 열의 값을 value 로 설정된 값으로 채워준다 (대표값 또는 예측값을 구한 후)

x = df_new['age'].mean()
y = df_new['score'].mean()
df_new['age'] = df_new['age'].fillna(value = x)
df_new['score'] = df_new['score'].fillna(value = y)

df_new
```

```python
#이상치 제거
#'A' class 를 3 class로 바꾸기

#개짱나지만 for문을 돌려야한다

for n in df.index : #data frame 의 index 속의 범위에서
    if df.loc[n, 'class'] == 'A' :  #index의 번호 속에 n을 넣고 n의 범위에서 class의 값을 찾는다
        df.loc[n, 'class'] = 3
        

df
        
```
