```python
df = pd.read_csv('data/pop_Seoul2.csv' , encoding = 'utf-8' )

df

print(df['구간'])
df
```

```python
plt.figure(figsize=(10,10))


year = df.loc[: , '구간']
seoul_pop = df.loc[: , '서울']
capital_pop = df.loc[: , '수도권']

x = df['구간']
y = seoul_pop

#x축에 년도, y축에 인구수가 있는 선 그래프 만들기

plt.plot(year , seoul_pop, color='green' , marker='*', linestyle='solid', \
         label='Seoul')
plt.plot(year , capital_pop, color='red' , marker='*', linestyle='solid', \
         label='Capital Region')

plt.xticks(range(12), year, rotation=45)

plt.title('Population Change')
plt.ylabel('(%)')

plt.legend()
plt.show()
```
