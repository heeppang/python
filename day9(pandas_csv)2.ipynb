```python
import pandas as pd


data = pd.read_csv("data/dirtydata.csv" , encoding = "utf-8")

data

new_data = data.copy()

new_data


```

```python
c = new_data['Calories'].mean()

new_data['Calories'] = new_data['Calories'].fillna(value = c)  #NA 값을 칼로리 평균치로 채워줬다

new_data
```

```python
y = pd.get_dummies(new_data['Duration'])   #.get_dummies = 'Duration' 컬럼의 평균값을 구해주는 함수

y

```

```python
bins = [new_data['Calories'].min(), new_data['Calories'].mean(), new_data['Calories'].max()]   #bins는 min 과 max 사이에 준 임의의 값임 . 칼로리를 평균값 기준으로 high low
                           
new_data['kcal']  = pd.cut(new_data['Calories'], bins=bins, labels = ['Low', 'High'])
new_data
new_data['Calories'].mean()
```

```python
bins = [new_data['Maxpulse'].min(), 120, new_data['Maxpulse'].max()]   #맥박을 120 기준으로 high low
                           
new_data['MaxP']  = pd.cut(x = new_data['Maxpulse'], bins=bins, labels = ['Low', 'High'])  # labels = [] 값은 내가 준 범위 구간보다 하나 적게 써야함
new_data
```
