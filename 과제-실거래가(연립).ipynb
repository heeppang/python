```python
import pandas as pd
import numpy as np
```

```python
price_2015 = pd.read_csv("data/연립다세대(매매)__실거래가_20220111113539.csv"  ,  encoding='cp949' , thousands=',' )
# ParserError: Error tokenizing data. C error: Expected 5 fields in line 3, saw 6  error 방지용으로 sep='\t' 삽입했음


# price_2015.drop(columns= ['거래유형','중개사소재지','해제사유발생일','도로명','건축년도','번지'] , inplace = True)
price_2015c = price_2015


bins = [price_2015c['전용면적(㎡)'].min(), 66.11, 99.16 , 132.22 , price_2015c['전용면적(㎡)'].max()]   #bins는 min 과 max 사이에 준 임의의 값임 . 칼로리를 평균값 기준으로 high low
                           
price_2015c['평수']  = pd.cut(price_2015c['전용면적(㎡)'] , bins=bins, labels = ['~ 20평', '20평대' , '30평대' , '40평 ~'])
price_2015c


under_20_2015 = price_2015c.loc[price_2015c['평수'] == '~ 20평']
under_30_2015 = price_2015c.loc[price_2015c['평수'] == '20평대']
under_40_2015 = price_2015c.loc[price_2015c['평수'] == '30평대']
upper_40_2015 = price_2015c.loc[price_2015c['평수'] == '40평 ~']

un_20 = under_20_2015['전용면적(㎡)'].mean()
un_30 = under_30_2015['전용면적(㎡)'].mean()
un_40 = under_40_2015['전용면적(㎡)'].mean()
up_40 = upper_40_2015['전용면적(㎡)'].mean()


avg_un_20_2015 = int((under_20_2015['거래금액(만원)'].mean()/un_20)*10000)
avg_un_30_2015 = int((under_30_2015['거래금액(만원)'].mean()/un_30)*10000)
avg_un_40_2015 = int((under_40_2015['거래금액(만원)'].mean()/un_40)*10000)
avg_up_40_2015 = int((upper_40_2015['거래금액(만원)'].mean()/up_40)*10000)






print('~ 20평 형대 ㎡ 당 평균 거래가는 ',format(avg_un_20_2015, ',d'),'원/㎡')
print('20평 형대 ㎡ 당 평균 거래가는 ',format(avg_un_30_2015, ',d'),'원/㎡')
print('30평 형대 ㎡ 당 평균 거래가는 ',format(avg_un_40_2015, ',d'),'원/㎡')
print('40평 형대 ㎡ 당 평균 거래가는 ',format(avg_up_40_2015, ',d'),'원/㎡')
```
