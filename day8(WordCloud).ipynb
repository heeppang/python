```python
mask = np.array(Image.open('./data/mask.png'))

wordcloud = WordCloud(max_font_size = 200,
                      stopwords = spwords,
                      background_color = '#000000',  #  #FFFFFF = white #000000 = black
                      # random_state = 50,
                      # colormap = 'plasma',  #matplotlib colormap 이용
                      mask = mask,
                      font_path = 'C:\Windows\Fonts\impact.ttf',
                      width = 1000,
                      height = 800).generate(text)
```
