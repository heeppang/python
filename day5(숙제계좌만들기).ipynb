```python
#Account class

'''
변수 2개 - 계좌, 잔액/ 멤버함수(메소드) 3개 - 입금,출금,송금
'''

class Account :
    account_cnt = 0   #class 변수
    
    
    def __init__(self, num) :
        
        self.__num = num   #self = instance method = 매개변수로 치지 않음
        self.__balance = 0
        #self.__balance = 0    <- python에선  __ 언더바 두개가 private 임. 호출하기 위해선 getter 함수를 만들고 변수로 호출한다

          
        
        Account.account_cnt = Account.account_cnt + 1   #class 명 Account. 을 붙혀 class 함수가 되어 class 당 으로 계산됨
        
        
        
    def set_num(self, num) :  #setter 함수는 getter 와 반대임. 입력값이 있고 해당 값을 self로 만든다
        self.__num = num
        
    def set_balance(self, balance) :
        self.__balance = balance
        
    def get_balance(other, balance) :
        return other.__balance
    
    
        
        
        
    
    #계좌번호
    def get_num(self) :
        return self.__num   #return에서 호출할 때도 똑같이 __ 언더바 두개를 붙혀줘야한다
    
    #잔액
    def get_balance(self) :
        return self.__balance
    
    #입금
    def deposit(self, money) :
        self.__balance = self.__balance + money
        return self.__balance    #balance 에 money를 더해줬으므로 밸런스 값을 return 한다
        
    #출금
    def withdraw(self, money) :
        if self.__balance >= money :
            self.__balance = self.__balance - money
            return('출금 후 잔액 %d 원' %self.__balance)    #balance 에 money를 더해줬으므로 밸런스 값을 return 한다
        else :
            return '잔액이 부족합니다'
        
        
    #송금
    def transfer(self, money, other) :  #송금할 금액과 받을 주체에 대한 변수가 필요
        self.__balance = self.__balance - money
        other.__balance = other.get_balance
        other.set_balance(other.__balance + money)
        other.__balance = other.__balanace + money
        
        print ('%d 송금 후 잔액 : %d 원 , 잔액 : %d 원' %(self.__num,self.__balance, other.__balance) )
        
        
        
    #정보출력
    def print_info(self) :
        print('계좌번호 %d번 잔액 : %d 원' % (self.__num , self.__balance))
        
    # #계좌 갯수 출력
    
    @classmethod
    def print_account_cnt(cls) :
        print(cls.account_count)
        
    
        
```
