```python
#Account class

'''
변수 2개 - 계좌, 잔액/ 멤버함수(메소드) 3개 - 입금,출금,송금
'''

class Account :
    account_cnt = 0   #class 변수
    
    
    def __init__(self, num) :
        
        self.__num = num   #self = instance method = 매개변수로 치지 않음
        self.__balance = 0
        #self.__balance = 0    <- python에선  __ 언더바 두개가 private 임. 호출하기 위해선 getter 함수를 만들고 변수로 호출한다

          
        
        Account.account_cnt = Account.account_cnt + 1   #class 명 Account. 을 붙혀 class 함수가 되어 class 당 으로 계산됨
        
        
        
    def set_num(self, num) :  #setter 함수는 getter 와 반대임. 입력값이 있고 해당 값을 self로 만든다
        self.__num = num
        
    def set_balance(self, balance) :
        self.__balance = balance
        
    def get_balance(other, balance) :
        return other.__balance
    
    
        
        
        
    
    #계좌번호
    def get_num(self) :
        return self.__num   #return에서 호출할 때도 똑같이 __ 언더바 두개를 붙혀줘야한다
    
    #잔액
    def get_balance(self) :
        return self.__balance
    
    #입금
    def deposit(self, money) :
        self.__balance = self.__balance + money
        return self.__balance    #balance 에 money를 더해줬으므로 밸런스 값을 return 한다
        
    #출금
    def withdraw(self, money) :
        if self.__balance >= money :
            self.__balance = self.__balance - money
            return('출금 후 잔액 %d 원' %self.__balance)    #balance 에 money를 더해줬으므로 밸런스 값을 return 한다
        else :
            return '잔액이 부족합니다'
        
        
    #송금
    def transfer(self, money, other) :  #송금할 금액과 받을 주체에 대한 변수가 필요
        self.__balance = self.__balance - money
        other.__balance = other.get_balance
        other.set_balance(other.__balance + money)
        other.__balance = other.__balanace + money
        
        print ('%d 송금 후 잔액 : %d 원 , 잔액 : %d 원' %(self.__num,self.__balance, other.__balance) )
        
        
        
    #정보출력
    def print_info(self) :
        print('계좌번호 %d번 잔액 : %d 원' % (self.__num , self.__balance))
        
    # #계좌 갯수 출력
    
    @classmethod
    def print_account_cnt(cls) :
        print(cls.account_count)
        
    
        
    
        
```

```python
'''
가로 세로
둘레 구하기 넓이 구하기
'''

class Rectangle :
    
    def __init__ (self, depth, width) :
        

        self.__depth = depth
        self.__width = width
        
    # def __repr__
    
    def __repr__(self) :   #representation
        return 'gg'
    
    def __str__(self) :
        return '가로 : %d 세로 : %d' %(self.__width , self.__depth)
    
    def get_depth(self) :     #getter 에는 return 이 무조건 세트
        return(self.__depth)
    
    def set_depth(self, depth) :  #setter에는 매개변수에 (self, ＊＊＊) 와 같은 형태로 써준다
        self.__depth = depth
    
    def get_width(self) :
        return(self.__width)
    
    def set_width(self, width) :
        self.__width = width
        
    
    def get_long(self) :   #매개변수가 아닌 멤버변수로 계산할 것이므로 self 옆에 다른 변수를 쓸 필요가 없음
        return (self.__depth+self.__width)*2
    
    def get_area(self) :
        return self.__depth * self.__width
        
    def __eq__(self , other_rect) :  #매개변수와 self를 비교하여 같으면 true 다르면 false를 printout 하는 함수
        
        return self.get_area() == other_rect.get_area()         # ex) r1 == r2 라고 쓰면 자동호출되는 함수
    
    def __gt__(self, other_rect) :     #greatter than : 크기를 비교하는 함수
        
        return self.get_area() > other_rect.get_area()         # ex) r1 > r2 라고 쓰면 자동호출되는 함수
    
    def __le__(self, other_rect) :     #less more : 크기를 비교하는 함수
        
        return self.get_area() > other_rect.get_area()   
     
    


        
    
        
#         self.depth = depth
#         self.width = width
        
    
#     def get_depth(self) :     #getter 에는 return 이 무조건 세트
#         return(self.depth)
    
#     def set_depth(self, depth) :  #setter에는 매개변수에 (self, ＊＊＊) 와 같은 형태로 써준다
#         self.depth = depth
    
#     def get_width(self) :
#         return(self.width)
    
#     def set_width(self, width) :
#         self.width = width
        
    
#     def get_long(self) :   #매개변수가 아닌 멤버변수로 계산할 것이므로 self 옆에 다른 변수를 쓸 필요가 없음
#         return (self.depth+self.width)*2
    
#     def get_area(self) :
#         return self.depth * self.width
```
