```python
import pandas as pd
```

```python
age =  pd.Series([20, 22, 24, 26, 28 ,30, 32, 34, 36 ,38])
score = pd.Series([2.4, 2.6, 3.0, 3.4, 3.7, 4.1, 4.3, 4.5, 3.9 , 4.2])
salary = pd.Series([2400, 2600 , 2700 , 2800 , 2900 , 3000 , 3200 , 3400 , 3800 , 4000])

print(len(age))
print(len(score))
print(len(salary))


# DataFrame 은 series 들을 합쳐놓은 것임

df = pd.DataFrame({ 'age' : age ,
                    'score' : score ,
                    'salary' : salary })


# print(df)

df

df.head() # 데이터 맨앞 5행만 확인

#df.head(3) # 데이터 맨앞 3행만 확인

df.tail() # 데이터 뒤쪽의 5행만 확인

df.loc[0]  # index 0번째의 모든 값을 출력
df.loc[[0,1]] # index 0,1번째의 값을 출력

df.index #RangeIndex(start=0, stop=11, step=1)

df.columns  #Index(['age', 'score', 'salary'], dtype='object')

df.values #값




df.describe()   # 요약 정보 확인 . 갯수 평균 등

df.T  #데이터의 전치

df.sort_values(by = 'score', ascending = True) 
#점수 기준 오름차순 정렬 . , ascending(오름차순) = True 는 default값으로 생략가능
df.sort_values(by = 'score', ascending = False) #점수 기준 내림차순 정렬


df.info() # 데이터의 전반적인 정보확인
```

```python
fruit = pd.DataFrame(
    {'Num' : [123, 456, 789, 1011, 1112],
     'Fruit' : ['Apple', 'Banana' , 'Cherry' , 'Lemon' , 'Peach']})
                
grade = pd.DataFrame(
    {'Num' : [123, 789, 1314],
     'Grade' : ['A' , 'B', 'C']})
                
price = pd.DataFrame({
    'No' : [123, 789, 1314],
    'Grade' : [100, 200, 500]})
```

```python
fruit
```

```python
grade
```

```python
price
```

```python
#df 합치는 방법. 유의미한 건 아니고 보통 data 양 늘릴 때 사용함

df = pd.concat([fruit,grade,price])

df
```

```python
#Inner Join

#두 df를 merge 했을 때 num값이 일치하는 것끼리만 merge하고 추출

df = pd.merge(fruit, grade, on = 'Num')
df
```

```python
#Left Join / Right Join / Outer Join

#Left Join = 두 column 중 왼쪽은 다 나오고 일치하는 부분이 있는 오른쪽 출력 (왼쪽을 기준으로 교집합)
#Right Join = 두 column 중 오른쪽은 다 나오고 일치하는 부분이 있는 왼쪽 출력 (오른쪽을 기준으로 교집합)

# df = pd.merge(fruit, grade, on =  'Num', how = 'left')
# df = pd.merge(fruit, grade, on =  'Num', how = 'right')
df = pd.merge(fruit, grade, on =  'Num', how = 'inner')
df

# df.isnull().sum()
```

```python
df = pd.merge(fruit , price , left_on = 'Num', right_on='No', how = 'left')
df

# df.isnull() #결측치
# df.isnull().sum()
```
