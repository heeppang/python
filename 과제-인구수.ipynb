```python
import pandas as pd

year      = pd.Series([2015,2016,2017,2018,2019,2020]) 
under_20 = pd.Series([117087,111875,105227,98591,92613,86417]) 
upper_20 = pd.Series([357081,350569,339564,330867,320771,311990]) 
upper_60 = pd.Series([100415,105137,109612,114294,119521,124630]) 



popul = pd.DataFrame ( {'연도': year, 
                     '0~19세' : under_20,
                     '20~60세' : upper_20,
                     '60세이상' : upper_60,
                     '계' : '' }
)

popul['계'] = popul.loc[: , '0~19세':'60세이상'].sum(axis=1)

popul

# popul['20~60세'].unique()
```
